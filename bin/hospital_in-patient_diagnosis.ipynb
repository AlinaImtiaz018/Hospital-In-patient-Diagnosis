# Hospital In-Patient Diagnosis

 In this project, we conduct an Exploratory Data Analysis (EDA) on hospital in-patients dataset focusing on the diseases they are diagnosed with initially from the year 2018 to 2022. The data is divided into different age groups spanning from 0 to 90 years, consisting the information about the average length of stay at the hospital. The research questions to be investigated with the hospital patients dataset are as follows:

#### 1. How does the prevalence of diseases vary between male and female? 

#### 2. What are the most common diseases diagnosed among patients in different age groups? Is there any significant difference in the average length of stay at the hospital across different age groups?

#### 3. Are there any notable trends or changes in disease prevalence over the five-year period from 2018 to 2022?


First, we will start by importing all the relevant libraries that we going to use in our EDA project.

```python
fig_1_to_18 = px.histogram(df_male, x='Diagnosis', y='1 to 18',
                            title="Diagnosis : Distribution of Diseases by Age Group (MALE) : Year 2018 to 2022 ",
                            height=500, color_discrete_sequence=['blue'])

fig_18_to_30 = px.histogram(df_male, x='Diagnosis', y='18 to 30', color_discrete_sequence=['green'])

fig_30_to_50 = px.histogram(df_male, x='Diagnosis', y='30 to 50', color_discrete_sequence=['red'])

fig_50_to_70 = px.histogram(df_male, x='Diagnosis', y='50 to 70', color_discrete_sequence=['yellow'])

fig_70_to_90 = px.histogram(df_male, x='Diagnosis', y='70 to 90', color_discrete_sequence=['orange'])

fig_1_to_18.update_traces(opacity=0.5)  
for data in fig_18_to_30.data:
    fig_1_to_18.add_trace(data)

fig_1_to_18.update_traces(opacity=0.5)  
for data in fig_30_to_50.data:
    fig_1_to_18.add_trace(data)


fig_1_to_18.update_traces(opacity=0.5)  
for data in fig_50_to_70.data:
    fig_1_to_18.add_trace(data)

fig_1_to_18.update_traces(opacity=0.5)  
for data in fig_70_to_90.data:
    fig_1_to_18.add_trace(data)


fig_1_to_18.update_xaxes(title_text='Diagnosis')
fig_1_to_18.update_yaxes(title_text='Length of Stay')
fig_1_to_18.show()
```

The above histogram shows the total amount of average length of stay for Males with different diseases based on thier age groups. Overall, according to out data, the reason of stay inside the hopisal for males is because of Malignant Neoplasms. ## Mean values of length of stay for Males and Females

```python
mean_values_male = df_male[['1 to 18', '18 to 30', '30 to 50', '50 to 70', '70 to 90']].mean()
mean_values_female = df_female[['1 to 18', '18 to 30', '30 to 50', '50 to 70', '70 to 90']].mean()
mean_df_male = pd.DataFrame({'Age Group': mean_values_male.index, 'Mean Value': mean_values_male.values, 'Gender': 'Male'})
mean_df_female = pd.DataFrame({'Age Group': mean_values_female.index, 'Mean Value': mean_values_female.values, 'Gender': 'Female'})
mean_df_combined = pd.concat([mean_df_male, mean_df_female], ignore_index=True)

fig = px.bar(mean_df_combined, x='Age Group', y='Mean Value', color='Gender', barmode='group',
             title='Mean Values of Age Groups for Males and Females',
             labels={'Mean Value': 'Mean Value', 'Age Group': 'Age Group', 'Gender': 'Gender'})

fig.update_layout(width=600,height=600)
fig.show()
```

The above bar plot shows the average length of stays of males and females according to their age groups. Here too, it can be seen that with an increase in the age, the length of the stay also increases. However, Females have a lower length of stay at the hospitals that that of Males for each age group. 

The avarage length of stay of males in the age group 1 to 18 is 4.53 days. For the age group 70 to 90, the value equals 8.46 days which is almost double. 
Simlarly from Females, as age increases, the nummber of days of stays increases.

## Pie Chart - Distribution in Percentages

```python
selected_disease = 'Infections'
df_male_filtered = df_male[df_male['Diagnosis'] == selected_disease]
df_female_filtered = df_female[df_female['Diagnosis'] == selected_disease]

age_group_counts_male = df_male_filtered[['1 to 18', '18 to 30', '30 to 50', '50 to 70', '70 to 90']].sum()
age_group_counts_female = df_female_filtered[['1 to 18', '18 to 30', '30 to 50', '50 to 70', '70 to 90']].sum()

fig = make_subplots(rows=1, cols=2, specs=[[{'type':'domain'}, {'type':'domain'}]],
                    subplot_titles=['Males', 'Females'])
fig.add_trace(go.Pie(labels=age_group_counts_male.index, values=age_group_counts_male.values, name='Male'), 
              row=1, col=1)
fig.add_trace(go.Pie(labels=age_group_counts_female.index, values=age_group_counts_female.values, name='Female'), 
              row=1, col=2)
fig.update_layout(title='Distribution by Age Group for Males and Females : Infections')
fig.show()
```

For diseases cased by infections, 30.4% of Females and 27.6% of Males in the age group 70 to 90 contribute to the highest length of stay when comparing to all other age groups.

```python
df_2018 = df1[(df1['Year'] == 2018)]
df_2022 = df1[(df1['Year'] == 2022)]
diseases = ['Infections', 'Malignant Neoplasms', 'Blood', 'Respiratory', 'Disgestive', 'Musculoskeletal', 'Nervous System', 'Metabolic', 'Circulatory']
figures = []

for disease in diseases:

    df_disease_2018 = df_2018[df_2018['Diagnosis'] == disease]
    df_disease_2022 = df_2022[df_2022['Diagnosis'] == disease]

    df_disease_melted_2018 = pd.melt(df_disease_2018, id_vars=['Diagnosis'], 
                                      value_vars=['1 to 18', '18 to 30', '30 to 50', '50 to 70', '70 to 90'], 
                                      var_name='Age Group', value_name='Count')
    
    df_disease_melted_2022 = pd.melt(df_disease_2022, id_vars=['Diagnosis'], 
                                      value_vars=['1 to 18', '18 to 30', '30 to 50', '50 to 70', '70 to 90'], 
                                      var_name='Age Group', value_name='Count')
    

    df_disease_combined = pd.concat([df_disease_melted_2018.assign(Year=2018), 
                                     df_disease_melted_2022.assign(Year=2022)])
    
    
    fig = px.box(df_disease_combined, x='Age Group', y='Count', color='Year', 
                 title=f"Age Group Distribution of {disease} for year 2018 and 2022")
    
   
    figures.append(fig)

for fig in figures:
    fig.show()
```

The above generated plots describe the distribution of diseases for the year 2018 and 2022 across different age groups. 
From the above plot of Infection disease, we can see that the length of stay rates were higher in 2018 compared to 2022, with a decrease observed in age groups 1 to 18, 18 to 30, and 70 to 90, and an increase in the 30 to 50 age group. While the interquartile range remains similar for the 50 to 70 age group, the maximum value suggests a higher prevalence in 2018.

From the Malignant Neoplasm diseases, we can see a decrease in the length of stay across all age groups from 1 to 90. However, there are outliers present in the age group 1 to 18, suggesting variability in the length of stay of malignant neoplasm within this age range. This pattern suggests a general trend of declining diagnoses of Malignant Neoplasm diseases across all age groups over the specified time period. The presence of outliers in the age group 1 to 18 indicates that while there is an overall decrease, there are instances of variability or exceptions within this specific age range.

From the Blood diseases boxplot, we can see that there is a decrease in blood disease among the age groups 30 to 50, 50 to 70, and 70 to 90, while the age group 1 to 18 shows an increase over the same period. The average length of stay for the age group 18 to 30 remains relatively stable across the years. This suggests evolving patterns in the prevalence of blood diseases across various age groups over time. The decrease in diagnoses among older age groups and the increase among younger individuals may reflect shifts in disease incidence or changes in diagnostic practices.

The boxplot for the respiratory diseases, suggests that the respiratory diseases across the age groups 1 to 18 and 18 to 30 has decreased over time, while the diseases increased for the age groups 30 to 50, 50 to 70, and 70 to 90. The presence of outliers indicates potential variability or exceptional cases within certain age groups, warranting further investigation into underlying factors contributing to these trends.

The disgestive diseases boxplot suggests there has been a significant decline in the length of stay over time where the age groups from 18 to 90 years saw a decline while it was a bit higher in the age groups 1 to 18, signifing that they are more prone to digestive diseases and need in-patient treatment as compared to the other age groups. Also, seen in the circulatory and metabolic diseases.

The musculoskeletal diseases has decrease over time as well for the age groups indicating that there may be some decline in the length of stay over the years suggesting that the treatments may have worked while the length of stay for the nervous system diseases has increased over time especially for the elderly.
